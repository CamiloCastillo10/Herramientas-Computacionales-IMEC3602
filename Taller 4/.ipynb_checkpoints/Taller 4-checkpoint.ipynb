Ejercicio 2

La FFT  usa una base de expasion de cosenos y senos que se extiende por el dominio completo de la funcion, logrando la aproximacion a la derivada glbal de la funcion dada. Para ejemplificar esto, en el texto se hace un ejemplo con la siguiente funcion:

$u(x)=sech(x)$ 

$\frac{du}{dt}=−sech(x)tanh(x) $

$\frac{d^2u}{dt^2}=sech(x)−2sech^3(x)$

```python
def sech(x):
    return 1/(cosh(x))
```

```python
def tanh(x):
    return sinh(x)/cosh(x)
```

```python
L=20
n=128
x2=linspace(-L/2,L/2,n+1)
x=x2[0:n]
dx=x[1]-x[0]
u=sech(x)
ut=fft.fft(u)
ls1=linspace(0,n/2-1,n/2)
ls2=linspace(-n/2,-1,n/2)
ls=array(concatenate((ls1,ls2)))
k =(2*pi/L)*ls

# calculo de derivadas con FFT
u=array(u)
ut1=k*ut*1j
ut2=-k*k*ut
u1=fft.ifft(ut1).real
u2=fft.ifft(ut2).real

u1exact=[]
u2exact=[]

for i, xi in enumerate(x):
    u1exact.append(-sech(xi)*tanh(xi))
    u2exact.append(sech(xi)-2*(sech(xi))**3)

#calculo de primera derivada por diferencias finitas
#segundo orden precision
ux=[]
aux=(-3*u[0]+4*u[1]-u[2])/(2*dx)#ux(0)
ux.append(aux)# se añade el elemente ux(0)
for j in range(1,n-1):
    ux.append((u[j+1]-u[j-1])/(2*dx))
    
ux.append((-3*u[n-1]+4*u[n-2]-u[n-3])/(2*dx) )#se añade el elemento ux(n)


#Cuarto orden de presicion
ux2=[]
ux2.append((-3*u[0]+4*u[1]-u[2])/(2*dx))# se añade elemento ux2(0)
ux2.append((-3*u[1]+4*u[2]-u[3])/(2*dx))# se añade elemento ux2(1)
for j in range(2,n-2):
    ux2.append((-u[j+2]+8*u[j+1]-8*u[j-1]+u[j-2])/(12*dx))
    
ux2.append((3*u[n-2]-4*u[n-3]+u[n-4])/(2*dx))#Se añade el elemente ux2(n-1)
ux2.append((3*u[n-1]-4*u[n-2]+u[n-3])/(2*dx))#Se añade el elemente ux2(n)


#graficas
figure()
plot(x,u,'r',label='$u(x)$')
plot(x,u1,'g',label='$u´_{fft}(x)$')
plot(x,u1exact,'go',label='$u´(x)$')
plot(x,u2,'b',label='$u´´_{fft}(x)$')
plot(x,u2exact,'bo',label='$u´´(x)$')
legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

figure()
subplot(3,1,1)
plot(x,u1exact,'ks-',label='$u´(x)$')
plot(x,u1,'k',label='$u´_{fft}(x)$')
plot(x,ux,'ko',label='Diferencias finitas 2do Orden')
plot(x,ux2,'k*',label='Diferencias finitas 4to Orden')
xlim(-1.20, -0.75)
ylim(0.47 ,0.5)
legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
show()

plt.subplot(3,1,2)
plot(x,u1exact,'ks-',label='$u´(x)$')
plot(x,u1,'kv',label='$u´_{fft}(x)$')
plot(x,ux,'ko',label='Diferencias finitas 2do Orden')
plot(x,ux2,'k*',label='Diferencias finitas 4to Orden')
xlim(-0.9376, -0.9374)
ylim(0.49848, 0.49850)
legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
show()
plt.subplot(3,1,3)
plot(x,u1exact,'ks-',label='$u´(x)$')
plot(x,u1,'kv',label='$u´_{fft}(x)$')
plot(x,ux,'ko',label='Diferencias finitas 2do Orden')
plot(x,ux2,'k*',label='Diferencias finitas 4to Orden')
xlim(-0.9376, -0.9374)
ylim(0.498487, 0.498488)
legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
show()
```

En las tres graficas de arriba podemos notar lo cercano que puede resultar la aproximacion por medio de FFT a la funcion real.

Filtro Gaussiano: Este filtro eliminia ciertas frecuencias de manera aleatoria en una distribucion normal, este filtro se define como:

$F(k)=exp(−\tau(k−k_0)^2)$

```python
def filtro(k,m):
    return exp(-0.2*(k-m)**2)
        
```

```python
f0=filtro(k,0)
unft=f0*unt
unf=fft.ifft(unft)

noise=10
ut=np.fft.fft(f0)
real=[]
ima=[]
for i in range(0,n):
    real.append(random.randn())
    ima.append(1j*random.randn())
real=np.array(real)
ima=np.array(ima)
unt=ut+noise*(real)+noise*(ima)
un =fft.ifft(unt)

subplot(3,1,1)
plot(fft.fftshift(k),abs(fftshift(unt))/max(abs(fftshift(unt))),'k', label='Ruido')
plot(fft.fftshift(k),abs(fftshift(filtro(k,0)))/max(abs(fftshift(filtro(k,0)))),'b', label='Filtro Gaussiano')
xlabel('wavenumber (k)')
ylabel('|ut|/max(|ut|)')
xlim(-25,25,5)
ylim(0,1)
legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
show()

subplot(3,1,2)
plot(fftshift(k),abs(fftshift(unft))/max(abs(fftshift(unft))),'k',label='Señal despues de filtrar')
xlabel('wavenumber (k)')
ylabel('|ut|/max(|ut|)')
legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
show()

subplot(3,1,3)
plot(t,abs(unf),'g', label='Espectro limite de deteccion')
plot(t,abs(u),'k',label='Señal ideal')
xlabel('time (t)')
ylabel('|u|')
ylim(0,1)

legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
show()
```

# Ejercicio personal

Para este ejemplo se filtrara la voz de la cancion Everlong (acustica) de Foo Fighters, la razon de la seleccion de esta cancion es debido a que es una cancion acustica, unicamente cuenta con voz y guitarra. El objetivo de este ejercicio es lograr filtrar el espectro de la voz para unicamente dejar el sonido de la guitarra.

Se hizo uso de:

- Scipy.fft
- Scipy.waves
- Butter pasabandas

```python
archivo ='Foo  Fighters - Everlong (Acoustic).wav'
samples,sound=waves.read(archivo)#rate samples

size=shape(sound)
L=len(size)
canals=1

if L>1:
    canals=size[1]
if canals>1:
    canal=0
    xcanal=sound[:,canal]
else:
    uncanal=sound
inicioCancion=0
finCancion=4*60+9
a=int(inicioCancion*samples)
b=int(finCancion*samples)#numero de muestras 
```

```python
def butter_bandpass(lowcut, highcut, fs, order=4):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a
```

```python
def espectro(x):
    n = len(x) # longitud de la señal
    k = arange(n)
    Fs= samples
    ll = n/Fs
    frq = k/ll#frecuencia
    
    for i in range(0,n):
        if i>n/2:
            frq[i]=0
            
    X = fft.fft(x)/n # fftnormalizada
    
    for i in range(0,n):
        if i>n/2:
           X[i]=0
```

Segun al siguiente pagina web:
https://www.hispasonic.com/reportajes/tabla-rango-frecuencias-instrumentos-musicales/39

El espectro de la voz se encuentra entre 87 y 392, mientras que el espectro de la guitarra acustica es de 82 y 988. podemos notar que el rango de la voz se encuentra dentro de la guitarra. por eso primero vamos a intentar aislar la cancion en el rango de la voz, luego en el rango de la guitarra y por ultimo en el rango de la guitarra que no incluye la voz.

```python
#82-988 para filtrar sol
b, a = butter_bandpass(87, 392, samples, order=3)
z = lfilter(b, a, sound[:,0])
#espectro(sound[:,0])
#espectro(z)
figure()
tx=linspace(0,finCancion,len(z))
subplot(211)
plot(tx,sound[:,0],'r',label='original')
plot(tx,z,'b',label='filtrada')
xlabel('Tiempo (s)')
ylabel('Amplitud')
legend(loc='upper left')
show()
```
